Tính tổng các trường hợp nhiễm ở các quốc gia
mutate tạo trường mới
distinct lọc quan sát trùng

```r
d3 <- df_clean %>% 
  group_by(countriesAndTerritories) %>%
  mutate(Sum_case_model = sum(cases)) %>%
  arrange(countriesAndTerritories, desc(Sum_case_model)) %>%
  select(countriesAndTerritories, Sum_case_model) %>%
  distinct()
  d3
```
